## NumPy useful functions

#### np.sort() -> it returns the sorted copy of an array

```python
# importing libraries
import numpy as np
```

```python
# creating 1 D array
a = np.random.randint(1,100,15)
a
```

```python
# creating 2 D array
b = np.random.randint(1,100,24).reshape(6,4)
b
```

#### Sorting in 1 D array

```python
np.sort(a)
```

#### Sorting in 2 D array

```python
np.sort(b)
```
